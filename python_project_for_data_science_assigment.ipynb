```python
import yfinance as yf

tesla = yf.Ticker("TSLA")
tesla_stock_data = tesla.history(period="max")
tesla_stock_data.reset_index(inplace=True)
tesla_stock_data.head()
```

```python
import pandas as pd
import requests

url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

response = requests.get(url, headers=headers)
tables = pd.read_html(response.text)
tesla_revenue = tables[1] 
tesla_revenue.columns = ['Date', 'Revenue']
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != '-']
tesla_revenue.head()
```

```python
gamestop = yf.Ticker("GME")
gme_stock_data = gamestop.history(period="max")
gme_stock_data.reset_index(inplace=True)
gme_stock_data.head()
```

```python
import pandas as pd
import requests

url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

response = requests.get(url, headers=headers)
tables = pd.read_html(response.text)
gme_revenue = tables[1]
gme_revenue.columns = ['Date', 'Revenue']
gme_revenue = gme_revenue[gme_revenue['Revenue'] != '-']
gme_revenue.head()
```

```python
import matplotlib.pyplot as plt

def make_graph(stock_data, stock, title):
    plt.figure(figsize=(14,6))
    plt.plot(stock_data['Date'], stock_data['Close'], label=f"{stock} Close Price")
    plt.xlabel('Date')
    plt.ylabel('Close Price USD ($)')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
make_graph(tesla_stock_data, 'TSLA', 'Tesla Stock Price History')
```

```python
make_graph(gme_stock_data, 'GME', 'GameStop Stock Price History')
```
